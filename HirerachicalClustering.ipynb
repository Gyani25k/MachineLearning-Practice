```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r'Mall_Customers.csv')
df=pd.DataFrame(data)
df.head()
```

```python
x=df.iloc[:,[3,4]].values
print(x)
```

```python
import scipy.cluster.hierarchy as sch
dendograms=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendograms')
plt.xlabel("Data Points")
plt.ylabel("Eculedian Distance")
```

```python
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_pred=model.fit_predict(x)
y_pred
```

```python
plt.scatter(x[y_pred==0,0],x[y_pred==0,1],c='red',s=100,label='Cluster 1')
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],c='blue',s=100,label='Cluster 2')
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],c='yellow',s=100,label='Cluster 3')
plt.scatter(x[y_pred==3,0],x[y_pred==3,1],c='green',s=100,label='Cluster 4')
plt.scatter(x[y_pred==4,0],x[y_pred==4,1],c='pink',s=100,label='Cluster 5')
plt.title("Cluster of Customers (5)")
plt.xlabel('Annual Income')
plt.ylabel('Spending Score (1-100)')
plt.legend()
```
